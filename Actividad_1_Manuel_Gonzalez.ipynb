```python
df = TRAINDS.loc[:,['AMT_CREDIT','AMT_GOODS_PRICE', 'AMT_ANNUITY']]
```

```python
n_components = 3
```

```python
MYPCA = PCA(n_components=n_components)
```

```python
reduced = MYPCA.fit_transform(df.loc[:,["AMT_CREDIT','AMT_GOODS_PRICE', 'AMT_ANNUITY"]])
```

```python
for i in range(0, n_components):

  df['PC' + str(i + 1)] = reduced[:, i]

display(df.head())
```
